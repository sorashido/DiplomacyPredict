```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(style='white', context='notebook', palette='Set2')

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_1 = pd.read_csv("../data/target/1901-1902.csv")
df_2 = pd.read_csv("../data/target/1903-1904.csv")
df_3 = pd.read_csv("../data/target/1905-1906.csv")
df_4 = pd.read_csv("../data/target/1907-1908.csv")
df_5 = pd.read_csv("../data/target/1909-1910.csv")
df_6 = pd.read_csv("../data/target/1911-1912.csv")
df_7 = pd.read_csv("../data/target/1913-1914.csv")
df_8 = pd.read_csv("../data/target/1915-1916.csv")
df_9 = pd.read_csv("../data/target/1917-1918.csv")
df_10 = pd.read_csv("../data/target/1919-1920.csv")

dataset = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10])
```

```python
# dataset = pd.read_csv("../dataset/online-data/dataset.csv")
dataset = dataset.drop(columns=["Unnamed: 0"])
dataset = dataset.replace( '\n', ':', regex=True)
dataset = dataset.replace( '\r', '', regex=True)
dataset = dataset.replace( 'Spring', 'SPR', regex=True)
dataset = dataset.replace( 'Fall', 'FAL', regex=True)
dataset = dataset.replace( 'Series\(\[\], Name: target, dtype: int64\)', '', regex=True)
dataset = dataset.replace( 'Name: target, dtype: int64', '', regex=True)
dataset = dataset.replace( 'army', '', regex=True)
dataset = dataset.replace( 'fleet', '', regex=True)
dataset = dataset.replace('', np.nan)
dataset = dataset.dropna(subset=['out2'])
```

```python
dataset = dataset.replace({'Edinburgh': 'EDI', 'Liverpool': 'LVP', 'London': 'LON', 'Marseilles': 'MAR', 'Paris': 'PAR', 'Brest': 'BRE',
       'Venice': 'VEN', 'Rome': 'ROM', 'Naples': 'NAP', 'Munich': 'MUN', 'Berlin': 'BER', 'Kiel': 'KIE', 'Vienna': 'VIE',
       'Trieste': 'TRI', 'Budapest': 'BUD', 'Constantinople': 'CON', 'Ankara': 'ANK', 'Smyrna': 'SMY',
       'Moscow': 'MOS', 'St. Petersburg \(South Coast\) ': 'STP', 'Warsaw': 'WAR', 'Sevastopol': 'SEV',
       'Norwegian Sea': 'NWG', 'Yorkshire': 'YOR', 'North Sea': 'NTH', 'Picardy': 'PIC',
       'Mid-Atlantic Ocean': 'MAO', 'Tyrrhenian Sea': 'TYS', 'Silesia': 'SIL', 'Holland': 'HOL',
       'Albania': 'ALB', 'Serbia': 'SER','Gulf of Bothnia': 'GOB', 'Ukraine': 'UKR', 'Norway': 'NWY', 'Belgium': 'BEL', 'Portugal': 'POR',
       'Piedmont': 'PIA', 'Tunis': 'TUN', 'Galicia': 'GAL', 'Greece': 'GRE', 'Black Sea': 'BLA', 'Finland': 'FIN',
       'Sweden': 'SWE', 'Rumania': 'RUM', 'Helgoland Bight': 'HEL', 'Burgundy': 'BUR',
       'English Channel': 'ENC', 'Tyrolia': 'TYR', 'Ionian Sea': 'ION', 'Denmark': 'DEN', 'Wales': 'WAL',
       'Irish Sea': 'IRI', 'Armenia': 'ARM', 'St. Petersburg \(North Coast\)': 'STP',
       'North Atlantic Ocean': 'NAO', 'Ruhr': 'RUH', 'Aegean Sea': 'AEG', 'Apulia': 'APU',
       'Skagerrack': 'SKA', 'nan': 'NAN', 'Gascony': 'GAS', 'Prussia': 'PRU', 'Barents Sea': 'BAR', 'Bohemia': 'BOH',
       'Adriatic Sea': 'ADR', 'Western Mediterranean': 'WMS', 'Tuscany': 'TUS', 'Baltic Sea': 'BAL',
       'Gulf of Lyons': 'GOL', 'Eastern Mediterranean': 'EMS', 'Spain \(North Coast\)': 'SPA', 
       'Syria': 'SYR', 'Livonia': 'LIV', 'Clyde': 'CLY', 'Spain \(South Coast\)': 'SPA', 'North Africa': 'NAF',
       'Bulgaria \(South Coast\)': 'BUL', 'Bulgaria \(East Coast\)':'BUL'}, regex=True)
dataset = dataset.replace({'Bulgaria': 'BUL', 'Spain': 'SPA', 'St. Petersburg': 'STP'}, regex=True)
```

```python
dataset = dataset.replace( ' ', '', regex=True)
dataset = dataset.replace( '\.\.', '', regex=True)
```
